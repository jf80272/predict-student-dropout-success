```python
# Imports
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load dataset
df = pd.read_csv("data.csv", sep=';')
```

```python
df.describe()
```

```python
print(df.isnull().sum())
```

```python
display(df["Target"].unique())
```

```python
sns.countplot(x="Target", data=df)
for p in plt.gca().patches:
    plt.gca().annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2, p.get_height()), ha='center', va='bottom')
plt.xlabel("Target Type")
plt.ylabel("Count")
plt.title("Count of People in Each Target Type")
plt.show()
```

```python
sns.countplot(x="Target", hue="Gender", data=df)
for p in plt.gca().patches:
    plt.gca().annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2, p.get_height()), ha='center', va='bottom')
plt.xlabel("Target Type")
plt.ylabel("Count")
plt.title("Gender Count for Each Target Type")
plt.legend(title="Gender", labels=["Female", "Male"])
plt.show()
```

```python
sns.boxplot(x="Target", y="Admission grade", data=df)
plt.xlabel("Target Type")
plt.ylabel("Prior Institution Grade")
plt.title("Comparison of Prior Institution Grade by Target Type")
plt.show()
```

```python
# Calculate correlation matrix including target, mapping Target values temporarily
target_map = {"Dropout": 0, "Graduate": 1, "Enrolled": 2}
df_corr = df.copy()
df_corr["Target"] = df_corr["Target"].map(target_map)
corr_matrix = df_corr.corr()

# Visualize correlations
plt.figure(figsize=(32, 24))
sns.heatmap(corr_matrix, annot=True, fmt='.3f', cmap='coolwarm', 
            center=0, square=True, linewidths=1,
            cbar_kws={'label': 'Correlation Coefficient'})
plt.title('Feature Correlation Matrix', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```
